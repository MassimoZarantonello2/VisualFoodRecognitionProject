```python
test_gt = pd.read_csv('ground_truth/val_info.csv', header=None, names=['image', 'label'])
test_dataset = ImageDataset(test_gt, './val_set_degraded', train=False, deprecated=True, dataset_size=1000)
```

```python
idx, label, confidence = em.predict(test_dataset)
```

```python
def get_accuracy(prediction, gt_labels, percentage=True):
    correct = 0
    for i in range(len(prediction)):
        if percentage:
            label_prediction = np.argmax(prediction[i])
        else:
            label_prediction = prediction[i]
            if label_prediction == gt_labels[i]:
                correct += 1

    return correct / len(prediction)
```

```python
print(get_accuracy(label, test_gt['label'].tolist(), percentage=False))
```
